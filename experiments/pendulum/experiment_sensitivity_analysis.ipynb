## По заданной $g(t)$ восстановим наблюдения

```python
a = 0.15    # Сильная кубическая нелинейность
b = -1.2    # Резкий "горб" в середине интервала
c = 2.0     # Быстрый линейный рост
d = 3.0     # Низкое стартовое значение (увеличивает чувствительность)
true_coefs = torch.tensor([d, c, b, a], dtype=torch.float32)
g = lambda t: a * t**3 + b * t**2 + c * t + d

L = 0.5
```

## Восстановим $g(t)$ по наблюдениям

Make dataset and dataloaders from generated trajectories

```python
# add g to inital conditions and trajectories
g0 = g(torch.tensor([0., ])).expand(num_trajectories, 1)
x0 = torch.concat([z0, g0], dim=1)

g_traj = g(t)

x = torch.concat(
    [trajectories, g_traj[None,:, None].expand(num_trajectories, -1, 1)],
    dim=2
)

dataset = TensorDataset(x0, x)
dataloader = DataLoader(dataset, batch_size=5, shuffle=True)
```

```python
# approximation for the g field
class GPolyPrime(nn.Module):
    def __init__(self, coef_init: torch.FloatTensor):
        """coef_init: in form of (a_0, a_1, ..., a_n)
        """
        super().__init__()
        self.poly_coef = nn.Parameter(coef_init)
        self.order = coef_init.numel() - 1

    def forward(self, t, g):
        # g arg is dummy, but contains batch dim
        t_feat = torch.tensor([t ** i for i in range(self.order + 1)]).to(t.device)
        return self.poly_coef.dot(t_feat).expand_as(g)
```

Будем варировать начальное приближение аппроксимирующего полинома для $g'$ так, чтобы контролировать $C$-норму отклонения от истинного $g'$. Варируем старший коэф. Несложно посчитать, какое будет рассхождение с истинным полином $g(t)$ (ниже это считается численно). Обучаемся небольшое кол-во эпох и смотрим на сходимость

```python
from vf import ODELightningModule, PendulumNodeVf

from torchdiffeq import odeint_adjoint as odeint
from lightning.pytorch.loggers import WandbLogger, TensorBoardLogger

models = []
deviation_norms = torch.linspace(0.5, 6, 5)
print("True coefs", true_coefs)

for deviation_norm in deviation_norms:
    print(f"Deviation = {deviation_norm:.2}")

    # construct deviated init point for g'(t)
    init_coefs = (true_coefs * torch.arange(0, true_coefs.numel()))[1:]
    d = init_coefs.numel() - 1
    init_coefs[-1] += deviation_norm / t_span[1] ** d

    print(init_coefs)

    # build vector fields and lightning module
    g_prime_poly = GPolyPrime(init_coefs)
    full_vf = PendulumNodeVf(g_prime_poly, L)
    model = ODELightningModule(full_vf, t, g_traj, lr=1e-4)

    # compute C_norm for |g(t) - approx_g(t)|
    g_pred = odeint(g_prime_poly, g_traj[0][None, ...], t, atol=1e-3).squeeze()
    print("C_norm", torch.max(torch.abs(g_pred - g_traj)))

    logger = TensorBoardLogger("logs", f"deviation_{deviation_norm:.2}")

    trainer = pl.Trainer(max_epochs=5, enable_progress_bar=True, logger=logger, log_every_n_steps=5)
    trainer.fit(model, dataloader)
    models.append(model)
```

```python
for dev, model in zip(deviation_norms, models):
    # predict for first 3 inital conditions
    with torch.no_grad():
        pred = model(x0[:3], t)
    
    # g(t)

    fig, ax = plt.subplots(figsize=(10, 6))

    ax.plot(t.numpy(), g_traj.numpy(), label="$g$")
    ax.plot(t.numpy(), pred[0, :, 2].numpy(), label="$g$_pred")

    # compute C_norm for |g(t) - approx_g(t)|
    print("C_norm", torch.max(torch.abs(g_traj - pred[0, :, 2])))

    ax.legend()
    ax.grid(True)
    ax.set_ylabel("$g$")
    ax.set_xlabel("$t$")
    ax.set_title(f"Init deviation = {dev:.2}")

    # Trajectories

    fig, ax = plt.subplots(figsize=(10, 6))

    ax.plot(pred[0, :, 0].numpy(), pred[0, :, 1].numpy(), label="pred")
    ax.plot(trajectories[0, :, 0].numpy(), trajectories[0, :, 1].numpy(), label="true")

    ax.legend()
    ax.grid(True)
    ax.set_title(f"Trajectory 0")
    ax.set_ylabel(r"$\omega$")
    ax.set_xlabel(r"$\phi$");
```
